```python
import scipy
import numpy as np
import matplotlib.pyplot as plt
import cmath 
```

```python
def R(z):
    return np.exp(z*1j)

def derivative_R(z):
    return scipy.misc.derivative(func=R, x0=z)

def f(z):
    return (R(z)/derivative_R(z))*1j
```

```python
gamma = np.zeros(100, dtype=complex)

gamma[0] = 10
h = 1

for i in range(1, len(gamma)):
    gamma[i] = gamma[i-1] + h*f(z=gamma[i-1])
```

```python
re_gamma = gamma.real
im_gamma = gamma.imag
plt.plot(re_gamma, im_gamma, 'go-')
plt.ylabel('Imaginary')
plt.xlabel('Real')
plt.title("gamma")
plt.show()


R_values = R(gamma)
re_R = R_values.real
im_R = R_values.imag
plt.plot(re_R, im_R, 'go-')
plt.ylabel('Imaginary')
plt.xlabel('Real')
plt.title("R(gamma)")
plt.show()
```
